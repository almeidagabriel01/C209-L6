```python
import cv2
from matplotlib import pyplot as plt

# Carregar imagens.
img = cv2.imread("image.jpg")
img2 = cv2.imread("image2.jpg")
img3 = cv2.imread("imageWithoutFace.jpg")

# A imagem  carregada em escala de cinza (para deteco de objetos) e RGB (para exibio).
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

img_gray2 = cv2.cvtColor(img2, cv2.COLOR_BGR2GRAY)
img_rgb2 = cv2.cvtColor(img2, cv2.COLOR_BGR2RGB)

img_gray3 = cv2.cvtColor(img3, cv2.COLOR_BGR2GRAY)
img_rgb3 = cv2.cvtColor(img3, cv2.COLOR_BGR2RGB)

# Aqui, estamos carregando o classificador Haar Cascade. Este  um modelo treinado que o OpenCV usa para detectar objetos. Neste caso, face_data.xml  um arquivo que contm os dados do classificador para detectar rostos de uma imagem.
face_data = cv2.CascadeClassifier('face_data.xml')

# A funo detectMultiScale  usada para detectar objetos na imagem. Ela retorna uma lista de retngulos onde acredita que encontrou o objeto.
found_face = face_data.detectMultiScale(img_gray, minSize=(20, 20))
found_face2 = face_data.detectMultiScale(img_gray2, minSize=(20, 20))
found_face3 = face_data.detectMultiScale(img_gray3, minSize=(20, 20))

# Se algum objeto for encontrado, desenhamos um retngulo ao redor dele na imagem RGB (para que possamos v-lo). O retngulo  desenhado com a funo rectangle do OpenCV.
if len(found_face) != 0:
    for (x, y, width, height) in found_face:
        cv2.rectangle(img_rgb, (x, y), (x + height, y + width), (50, 255, 255), 5)
        
if len(found_face2) != 0:
    for (x, y, width, height) in found_face2:
        cv2.rectangle(img_rgb2, (x, y), (x + height, y + width), (50, 255, 255), 5)
        
if len(found_face3) != 0:
    for (x, y, width, height) in found_face3:
        cv2.rectangle(img_rgb3, (x, y), (x + height, y + width), (50, 255, 255), 5)

# Exibir imagem
plt.figure(figsize=(13, 13))
plt.subplot(1, 3, 1)
plt.imshow(img_rgb)
plt.subplot(1, 3, 2)
plt.imshow(img_rgb2)
plt.subplot(1, 3, 3)
plt.imshow(img_rgb3)
```
